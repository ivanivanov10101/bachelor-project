```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# pd.set_option('display.float_format', lambda x: '%.9f' % x)
```

```python
df = pd.read_csv('main.csv')
```

```python
def byte_to_mb(x):
    return x / 1e+6

data = df['Bytes']
data10k = df['Bytes'][0:10000]
data20k = df['Bytes'][10000:20000]
data30k = df['Bytes'][20000:30000]
data40k = df['Bytes'][30000:40000]
data50k = df['Bytes'][40000:50000]

byte_to_mb(data)
byte_to_mb(data10k)
byte_to_mb(data20k)
byte_to_mb(data30k)
byte_to_mb(data40k)
byte_to_mb(data50k)

data = data.to_list()
data10k = data10k.to_list()
data20k = data20k.to_list()
data30k = data30k.to_list()
data40k = data40k.to_list()
data50k = data50k.to_list()

fig, axs = plt.subplots(2, 3, figsize=(18, 18))
fig.suptitle('Bytes, Entire Dataset vs. set in steps of 10,000')
axs[0][0].plot(data, color='crimson')
axs[0][0].set_title('Entire Dataset')

axs[0][1].plot(data10k)
axs[0][1].set_title('First 10000')

axs[0][2].plot(data20k)
axs[0][2].set_title('First 20000')

axs[1][0].plot(data30k)
axs[1][0].set_title('First 30000')

axs[1][1].plot(data40k)
axs[1][1].set_title('First 40000')

axs[1][2].plot(data50k)
axs[1][2].set_title('First 50000')

for ax in axs.flat:
    ax.set(xlabel='Bytes', ylabel='Count')
```
